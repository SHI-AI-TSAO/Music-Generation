```
import numpy as np 
from matplotlib import pyplot as plt 
```

```
f=open("/content/drive/My Drive/進行曲.txt","r")
lines=f.readlines()
accuracy=[]
val_accuracy=[]
loss = []
val_loss = []
i = 0
for x in lines:
  if (i%2 == 1):
    accuracy.append(float(x.split(' ')[10]))
    loss.append(float(x.split(' ')[7]))
    val_accuracy.append(float(x.split(' ')[16][0:6]))
    val_loss.append(float(x.split(' ')[13][0:6]))
  i += 1
print(loss)
print(val_loss)
f.close()
```

```
x = np.arange(1,len(accuracy)+1)
fig, ax1 = plt.subplots(dpi=350)
ax2 = ax1.twinx()
plt.title("Accuracy[solid] and Loss[dash] with epoch\nMarching Music")
plt.xlabel("Epoch")
plt.xlim([1,21])
plt.xticks([4,9,14,19],["5","10","15","20"])
#plt.figure(dpi=350)
ax1.plot(x,accuracy,'navy')
ax1.plot(x,val_accuracy,"salmon")
ax2.plot(x,loss,"--",color="navy")
ax2.plot(x,val_loss,"--",color="salmon")
ax1.set_ylim([0,1.05])
ax1.set_ylabel("Accuracy(%)")
#ax2.set_ylim([0,1.05])
ax2.set_ylabel("Loss")
ax1.legend(["Training Data","Validation Data"],loc = "center right")
ax1.grid()
plt.savefig("/content/drive/My Drive/marching")
```
